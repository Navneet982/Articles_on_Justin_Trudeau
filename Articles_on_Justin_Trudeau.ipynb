## Evolution of the "No. of articles" from 2018 till today

```python
df['Year'] = pd.DatetimeIndex(df['Date']).year
```

```python
df
```

```python
Year= df["Year"].value_counts()
Year
```

```python
import matplotlib.pyplot as plt

```

```python
df['Year'].value_counts().sort_index().plot.barh()
```

```python
Section = ["World news", "Opinion" , "US news", "Politics"]
```

```python
# x-axis values
x = Section
# y-axis values
y = Year
  
# plotting points as a scatter plot
plt.scatter(x, y, label= "Numbers", color= "blue", 
            marker= ".", s=250)
  
# x-axis label
plt.xlabel('Most frequent article types')
# frequency label
plt.ylabel('Number of articles')
# plot title
plt.title('Evolution overview')
# showing legend
plt.legend()
  
# function to show the plot
plt.show()
```
